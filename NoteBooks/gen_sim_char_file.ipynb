```python
SOFTMAX_CSV_PATH = os.path.join('..', 'FOCR', 'FontOCR', 'output', 'softmax_only_lower', 'all.csv')
SIM_CHAR_FILE_PATH = os.path.join('only_lower_worst_and_sim_char.csv')
all_softmax_df = pd.read_csv(SOFTMAX_CSV_PATH)
all_softmax_df = all_softmax_df.set_index(["FontName", "ActualChar"])
SIM_NUM = 4
SIM_COL_NAME = list(map(lambda s: 'sim'+str(s), range(SIM_NUM)))
```

```python
def extract_diagonal_component(df):
  acc_dict = {}
  for i in range(len(string.ascii_lowercase)):
    acc_dict[string.ascii_lowercase[i]] = df.iloc[i, i%len(string.ascii_lowercase)]
  return acc_dict
```

```python
columns = ["FontName", "ActualChar", *SIM_COL_NAME]
sim_char_df = pd.DataFrame(index=[], columns=columns)

for font_name in all_softmax_df.index.get_level_values('FontName').drop_duplicates():
  font_softmax_df = all_softmax_df.xs(font_name, level=0)
  diagonal_component = extract_diagonal_component(font_softmax_df)
  worst_char = sorted(diagonal_component.items(), key=lambda x: x[1])[0][0]
  shim_chars = font_softmax_df.loc[worst_char, :].nlargest(2).index
  if (shim_chars[0] != worst_char):
    shim_char = shim_chars[0]
  else:
    shim_char = shim_chars[1]
  sims = [worst_char, worst_char.upper(), shim_char, shim_char.upper()]


  for char in all_softmax_df.index.get_level_values('ActualChar').drop_duplicates():
    df = pd.DataFrame([[font_name, char, *sims]], columns=columns)
    sim_char_df = pd.concat([sim_char_df, df], ignore_index=True)
```
